```python
# Import Dependencies
import json
import requests
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager

# Set the executable path
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit the Mars NASA news site
url = 'https://redplanetscience.com'
browser.visit(url)
# Optional delay for loading the page
browser.is_element_present_by_css('div.list_text', wait_time=1)
# Set up the HTML parser
html = browser.html
news_soup = soup(html, 'html.parser')
```

```python
# Preview HTML code
print(news_soup)
```

```python
# Parent element of each article
slide_elem = news_soup.find_all('div', class_='list_text')
```

```python
# Scrape of the title
title_elem = news_soup.find_all('div', class_='content_title')

a=1
for title in title_elem:
    print(a, ".", title.text)
    a=a+1
```

```python
# Scrape of the preview text
news_p = news_soup.find_all('div', class_='article_teaser_body')

b=1
for news in news_p:
    print(b, ".", news.text)
    b=b+1
```

```python
# Store each title-and-preview pair in a Python dictionary. And, give each dictionary two keys: title and preview
# Store all the dictionaries in a Python list
mars_list = []
for elem in slide_elem:
    mars_data = {
        "title": "",
        "preview": ""
    }
    title = elem.find('div',class_='content_title').text
    preview = elem.find('div', class_='article_teaser_body').text
    mars_data["title"] = title
    mars_data["preview"] = preview
    mars_list.append(mars_data)
# Print the list
print(mars_list)
```

```python
# Store the scraped data in a file or database. Export the scraped data to JSON file.
json_mars = json.dumps(mars_list)
print(json_mars)
with open("json_mars.json", "w") as my_file:
    my_file.write(json_mars)
```

```python
# Quit browser 
browser.quit()
```
